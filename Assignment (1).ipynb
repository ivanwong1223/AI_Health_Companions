Why Med-BERT?
Handles medical text  Ideal for chatbot symptom classification.
Learns patient health patterns  Good for disease risk prediction.
Pretrained on real patient data (EHRs)  More reliable than general NLP models like GPT.

```python
from transformers import AutoTokenizer, AutoModelForCausalLM, AutoModelForSequenceClassification
import torch

# Load Med-BERT for disease classification
medbert_model_name = "GerMedBERT/medbert-512"  # Replace with actual model path
medbert_tokenizer = AutoTokenizer.from_pretrained(medbert_model_name)
medbert_model = AutoModelForSequenceClassification.from_pretrained(medbert_model_name)

# Load Falcon-7B for chatbot responses
chatbot_model_name = "tiiuae/falcon-7b-instruct"
chatbot_tokenizer = AutoTokenizer.from_pretrained(chatbot_model_name)
chatbot_model = AutoModelForCausalLM.from_pretrained(chatbot_model_name)

# Function for disease classification
def classify_disease(text):
    inputs = medbert_tokenizer(text, return_tensors="pt")
    outputs = medbert_model(**inputs)
    predicted_class = torch.argmax(outputs.logits).item()
    return predicted_class

# Function for chatbot responses
def chatbot_response(user_input):
    inputs = chatbot_tokenizer(user_input, return_tensors="pt")
    outputs = chatbot_model.generate(**inputs, max_length=100)
    return chatbot_tokenizer.decode(outputs[0], skip_special_tokens=True)

# Interactive chatbot loop
while True:
    user_input = input("You: ")
    if user_input.lower() in ["exit", "quit"]:
        break

    disease_class = classify_disease(user_input)
    print(f"Classified disease category: {disease_class}")

    chatbot_reply = chatbot_response(user_input)
    print("AI:", chatbot_reply)
```

```python
import PySimpleGUI as sg
from transformers import pipeline

# Load the question answering pipeline
model_name = "bert-large-uncased-whole-word-masking-finetuned-squad"
qa_pipeline = pipeline("question-answering", model=model_name, tokenizer=model_name)

# Initialize conversation history dictionary
conversation_history = {
    "context": None,
    "questions": [],
    "answers": []
}

def answer_question(context, question):
    try:
        if len(conversation_history['questions']) != 0:
            concatenated_text = ""
            for history_question, answer in zip(conversation_history["questions"], conversation_history["answers"]):
                concatenated_text += f"{history_question} {answer}. "
            
            question = concatenated_text + question


        result = qa_pipeline(question=question, context=context)
        return question, result['answer'], result['score']
    except Exception as e:
        print("Error occurred while answering the question:", e)
        return "Sorry, I couldn't find an answer to that question."



def add_to_conversation_history(context, question, answer):
    # Add the current context, question, and answer to conversation history
    conversation_history["context"] = context
    conversation_history["questions"].append(question)
    conversation_history["answers"].append(answer)

# def answer_with_history(question):

#         # If it's a follow-up question, use the context and previous answers to answer
#         context = conversation_history["context"]
#         questions = conversation_history["questions"]
#         answers = conversation_history["answers"]
#         return answer_question(context, question)
#         # If it's not a follow-up question, answer it directly

# Define GUI layout
layout = [
    [sg.Text("Question-Answering System")],
    [sg.Text("Context:"), sg.Multiline(key='-CONTEXT-', size=(60, 10))],
    [sg.Multiline(size=(60, 10), key='-OUTPUT-', disabled=True)],
    [sg.InputText(key='-INPUT-')],
    [sg.Button('Ask'), sg.Button('Exit')],
]

# Create the GUI window
window = sg.Window('QA System', layout)

# Main event loop
while True:
    event, values = window.read()

    if event == sg.WINDOW_CLOSED or event == 'Exit':
        break

    user_input = values['-INPUT-']
    context = values['-CONTEXT-']

    if user_input:
        question = user_input

        # Get response to the user's question
        concat, answer, score = answer_question(context, question)
        add_to_conversation_history(context, question, answer)


        # Display the answer in the GUI window
        window['-OUTPUT-'].print("Question: " + concat)
        window['-OUTPUT-'].print("Answer: " + answer)
        window['-OUTPUT-'].print("Score: " + str(score))
        window['-OUTPUT-'].print("")

window.close()
```

```python
import PySimpleGUI as sg
from transformers import pipeline
import PyPDF2
import spacy

# Initialize conversation history dictionary
conversation_history = {
    "context": None,
    "questions": [],
    "answers": []
}

def answer_question_roberta(context, question):
    qa_pipeline = pipeline("question-answering", model="deepset/roberta-base-squad2", tokenizer="deepset/roberta-base-squad2")
    try:
        if len(conversation_history['questions']) != 0:
            concatenated_text = ""
            for history_question, answer in zip(conversation_history["questions"], conversation_history["answers"]):
                concatenated_text += f"{history_question} {answer}. "
            
            question = concatenated_text + question

        result = qa_pipeline(question=question, context=context)
        return question, result['answer'], result['score']
    except Exception as e:
        print("Error occurred while answering the question with RoBERTa:", e)
        return "Sorry, I couldn't find an answer to that question with RoBERTa."

def answer_question_bert(context, question):
    qa_pipeline = pipeline("question-answering", model="bert-large-uncased-whole-word-masking-finetuned-squad", tokenizer="bert-large-uncased-whole-word-masking-finetuned-squad")
    try:
        if len(conversation_history['questions']) != 0:
            concatenated_text = ""
            last_question = conversation_history["questions"][-1]
            last_answer = conversation_history["answers"][-1]
            concatenated_text = f"{last_question} {last_answer}"
            question = concatenated_text + question

            coref = spacy.load("en_core_web_sm")
            coref.add_pipe("fastcoref")

            doc = coref(question, component_cfg={"fastcoref": {'resolve_text': True}})
            resolved_text = doc._.resolved_text
            resolved_text = resolved_text.replace(f"{last_question} {last_answer}", "").strip()
            question = resolved_text.strip()

        result = qa_pipeline(question=question, context=context)
        return question, result['answer'], result['score']
    except Exception as e:
        print("Error occurred while answering the question with BERT:", e)
        return "Sorry, I couldn't find an answer to that question with BERT."

# Define GUI layout
layout = [
    [sg.Text("Question-Answering System")],
    [sg.Text("PDF File:"), sg.InputText(key='-PDF_FILE-'), sg.FileBrowse()],
    [sg.Text("Context:"), sg.Multiline(key='-CONTEXT-', size=(60, 10))],
    [sg.Multiline(size=(60, 10), key='-OUTPUT-', disabled=True)],
    [sg.InputText(key='-INPUT-')],
    [sg.Text("Select Model:"), sg.Combo(['RoBERTa', 'BERT'], key="-MODEL-", default_value="RoBERTa")],
    [sg.Button('Load PDF'), sg.Button('Ask'), sg.Button('Exit')],
]

# Create the GUI window
window = sg.Window('QA System', layout)

# Main event loop
while True:
    event, values = window.read()

    if event == sg.WINDOW_CLOSED or event == 'Exit':
        break

    if event == 'Load PDF':
        pdf_file_path = values['-PDF_FILE-']
        if pdf_file_path:
            with open(pdf_file_path, 'rb') as file:
                pdf_reader = PyPDF2.PdfReader(file)
                context = ""
                for page in pdf_reader.pages:
                    context += page.extract_text()
            window['-CONTEXT-'].update(context)

    if event == 'Ask':
        user_input = values['-INPUT-']
        context = values['-CONTEXT-']
        selected_model = values["-MODEL-"]

        if user_input:
            question = user_input

            if selected_model == "RoBERTa":
                concat, answer, score = answer_question_roberta(context, question)
            elif selected_model == "BERT":
                concat, answer, score = answer_question_bert(context, question)

            conversation_history["context"] = context
            conversation_history["questions"].append(question)
            conversation_history["answers"].append(answer)

            # Display the answer in the GUI window
            window['-OUTPUT-'].print("Question: " + concat)
            window['-OUTPUT-'].print("Answer: " + answer)
            window['-OUTPUT-'].print("Score: " + str(score))
            window['-OUTPUT-'].print("")

window.close()
```

```python
from sentence_transformers import SentenceTransformer, util

# Load a sentence transformer model
model = SentenceTransformer('paraphrase-MiniLM-L6-v2')

# Symptoms database
symptoms_db = [
    ("Fever, Cough, Sore Throat", "Flu"),
    ("Fever, Headache, Muscle Pain", "Dengue"),
    ("Cough, Shortness of Breath, Chest Pain", "Pneumonia"),
    ("Fatigue, Weight Loss, Frequent Urination", "Diabetes")
]

# Precompute embeddings
symptom_texts = [s[0] for s in symptoms_db]
symptom_embeddings = model.encode(symptom_texts)

def find_disease(user_input):
    user_embedding = model.encode(user_input)
    scores = util.cos_sim(user_embedding, symptom_embeddings)[0]
    
    best_match_idx = scores.argmax()
    confidence = scores[best_match_idx].item()

    if confidence > 0.7:  # Confidence threshold
        return f"You might have {symptoms_db[best_match_idx][1]}. Please take care."
    else:
        return "I'm unsure about your symptoms. Please consult a doctor."

# Test
user_input = input("Enter your symptoms: ")
print(find_disease(user_input))
```
